# Setup

```python
# pip install python-dotenv openai
from dotenv import load_dotenv
import os
from openai import OpenAI
```

```python
load_dotenv(dotenv_path='secrets.env')
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

# Embedding Queries

```python
query_1 = "a short article about the early days of Google"
```

```python
embedding_ada_002 = client.embeddings.create(
    model="text-embedding-ada-002", # 1536 dims
    input=[query_1]
).data[0].embedding

print(len(embedding_ada_002))
print(embedding_ada_002)

embedding_3_small = client.embeddings.create(
    model="text-embedding-3-small", # 1536 dims
    input=[query_1]
).data[0].embedding

print(len(embedding_3_small))
print(embedding_3_small)

embedding_3_large = client.embeddings.create(
    model="text-embedding-3-large", # 3072 dims
    input=[query_1]
).data[0].embedding

print(len(embedding_3_large))
print(embedding_3_large)
```
